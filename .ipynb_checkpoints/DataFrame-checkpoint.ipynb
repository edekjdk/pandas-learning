## Sorting data

```python
import pandas as pd
import numpy as np

fuel = pd.read_csv("./data/fuel.csv", usecols = ['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'], index_col = "Vehicle ID")
fuel.head()
```

```python
fuel.loc[27681, 'Combined MPG (FT1)'] = np.NaN
```

```python
fuel.head()
```

```python
fuel.sort_values(by=["Make", "Model"]).head()
```

```python
fuel.sort_values(ascending=False, by="Combined MPG (FT1)").head()
```

```python
fuel.sort_values(by=["Make", "Combined MPG (FT1)"], ascending=[True, False]).head()
```

```python
fuel.sort_values(by="Combined MPG (FT1)", na_position="first").head()
```

```python
fuel.sort_values(by="Combined MPG (FT1)", na_position="first", inplace=True)
fuel
```

```python
fuel.sort_index().head()
```
