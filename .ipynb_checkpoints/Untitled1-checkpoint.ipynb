```python
df = df[[
    #'video_id', 
    'title', 'description', 'published_date', 
    #'channel_id',
    'channel_title', 'tags', 
    #'category_id', 
    'view_count', 'like_count',
    'comment_count', 'duration' 
    #'thumbnail'
]].copy()
```

```python
df.published_date = pd.to_datetime(df.published_date)
```

```python
df
```

```python
df['duration'] = pd.to_timedelta(df['duration'])
```

```python
df['duration'] = df['duration'].apply(lambda x: f"{int(x.total_seconds()//3600):02}:{int((x.total_seconds()%3600)//60):02}:{int(x.total_seconds()%60):02}")
```

```python
df.loc[df.duration.idxmax(), ["title", "duration"]]
```

```python
df.isna().sum()
```

```python
df = df.loc[~df.duplicated(subset = ["title", "published_date","duration"])].reset_index(drop=True).copy()
```

```python
df.duplicated().sum()
```

```python
df.published_date.dt.year.value_counts()
```

```python
import matplotlib.ticker as ticker
df.plot(kind = "scatter", x = "duration", y = "view_count", alpha = 0.5)
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.show()
```

```python
df.view_count.plot(kind = "hist", bins=30)
```

```python
df.view_count.sort_values(ascending = False).head(30)
```

```python
below50m = df.loc[df.view_count < 10_000_000]
```

```python
below50m.plot(kind = "scatter", x = "duration", y = "view_count", alpha = 0.5)
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.show()
```
