```python
import pandas as pd
```

```python
ds_salaries = pd.read_csv("./data/DataScience_salaries_2025.csv", usecols = ["work_year", "experience_level", "job_title", "salary_in_usd","company_location", "company_size" ])
```

```python
ds_salaries
```

```python
ds_salaries.monthly_salary = ds_salaries.monthly_salary.round(2)
```

```python
mean_salary = ds_salaries.monthly_salary.mean().round(2)
mean_salary
```

```python
max_salary = ds_salaries.monthly_salary.max()
max_salary
```

```python
ds_salaries.company_location.value_counts()
```

```python
ds_pol = ds_salaries.query("company_location == 'PL'")
ds_us = ds_salaries.query("company_location == 'US'")
```

```python
us_mean_salary = ds_us.salary_in_usd.mean().round(2)
pol_mean_salary = ds_pol.salary_in_usd.mean().round(2)

print("Mean salary in USA: {}, mean salary in Poland: {}".format(us_mean_salary, pol_mean_salary))
```

```python
percentage = round((us_mean_salary - pol_mean_salary)/pol_mean_salary * 100,2)
```

```python
print("{}%".format(percentage))
```
