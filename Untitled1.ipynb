```python
df = df[[
    #'video_id', 
    'title', 'description', 'published_date', 
    #'channel_id',
    'channel_title', 'tags', 
    #'category_id', 
    'view_count', 'like_count',
    'comment_count', 'duration' 
    #'thumbnail'
]].copy()
```

```python
df.published_date = pd.to_datetime(df.published_date)
```

```python
df
```

```python
df['duration'] = pd.to_timedelta(df['duration'])
```

```python
df['duration'] = df['duration'].apply(lambda x: f"{int(x.total_seconds()//3600):02}:{int((x.total_seconds()%3600)//60):02}:{int(x.total_seconds()%60):02}")
```

```python
df.loc[df.duration.idxmax(), ["title", "duration"]]
```

```python
df = df.loc[~df.duplicated(subset = ["title", "published_date","duration"])].reset_index(drop=True).copy()
```
