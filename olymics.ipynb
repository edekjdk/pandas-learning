```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

## Creating DataFrames

```python
df_athlete = pd.read_csv("./data/olympics/Olympic_Athlete_Biography.csv")
events = pd.read_csv("./data/olympics/Olympic_Athlete_Event_Details.csv")
```

```python
df = pd.merge(df_athlete, events, on = "athlete_id")
df.head()
```

```python
df.columns
```

```python
df = df[[
        #'athlete_id', 
        'name', 'sex', 'born', 'height', 'weight', 'country',
        'country_noc_x', 
        #'description', 'special_notes', 
        'edition',
        #'edition_id', 'country_noc_y', 
        'sport', 'event', 
        #'result_id', 'athlete',
       'pos', 'medal', 'isTeamSport']]
df.head()
```

```python
df = df.rename(columns = {
                    "born" : "born_date",
                    "country_noc_x" : "country_sym",
                    "isTeamSport" : "is_team_sport"
})
```

```python
df = df[~df.duplicated(subset = ["name", "edition", "event"])]
df
```

```python
df.pos = df.pos.str.replace("=", "")
df.medal = df.medal.fillna("no")
df.medal = df.medal.str.capitalize()
df
```
